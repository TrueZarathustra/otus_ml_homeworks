### Searching for correlation

```python
data_corr = data.loc[:, data.columns!='id'].corr() 
display(data_corr)
```

```python
# Нарисуем heatmap для матрицы корреляций

fig = plt.figure(figsize=(9.5,8))
ax = sns.heatmap(
    data_corr, 
    vmin=-1, vmax=1, center=0,
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
ax.set_yticklabels(
    ax.get_yticklabels(),
    rotation=0,
)
ax.plot()
```

```python
# Найдем параметры с высокой корреляцией

CORRELATION_THRESHOLD = 0.90

# Stack-overflow-driven-development to remove duplicates from correlation matrix
corr_list = data_corr[abs(data_corr) >= 0.01].stack().reset_index()
corr_list = corr_list[corr_list['level_0'].astype(str)!=corr_list['level_1'].astype(str)]
# filtering out lower/upper triangular duplicates 
corr_list['ordered-cols'] = corr_list.apply(lambda x: '-'.join(sorted([x['level_0'],x['level_1']])),axis=1)
corr_list = corr_list.drop_duplicates(['ordered-cols'])
corr_list.drop(['ordered-cols'], axis=1, inplace=True)

# rename and sort to find strongly correlated pairs
corr_list.columns = ['param_1', 'param_2', 'correlation']
corr_list['correlation'] = corr_list['correlation'].abs()
corr_list = corr_list.sort_values(by=['correlation'], ascending=False)

display(corr_list[corr_list['correlation'] > CORRELATION_THRESHOLD])
```

### kNN with default parameters

```python
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, roc_curve, roc_auc_score, auc
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


data = pd.read_csv('data/data.csv')
data = data.drop('Unnamed: 32', axis=1)
```

```python
# Подготовим данные для обучения модели
diag = data.diagnosis.replace('M', 1).replace('B', 0)
# diag = data.diagnosis
data = data.drop(['id', 'diagnosis'], axis = 1 )
```

```python
# Разбиваем на train-test выборки в соотношении 70 на 30
X_train, X_test, y_train, y_test = train_test_split(data, diag, test_size=0.3, random_state=37)

print('Shape of training set is {0},{1}\nShape of testing set is {2},{3}'.format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))

display(X_train.head(3), y_train.head(3))

# Нормализуем данные, т.к. сейчас их масштаб не соразмерен друг другу: есть параметры со средним значением - сотни, 
# а есть, со средним значением на 4 порядка меньше
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
def quality(prediction_y, true_y):
    accuracy = accuracy_score(prediction_y, true_y)
    precision = precision_score(prediction_y, true_y)
    recall = recall_score(prediction_y, true_y)
    f1 = f1_score(prediction_y, true_y)
    print("Accuracy:  {:.3f}\nPrecision: {:.3f}\nRecall:    {:.3f}\nF1-score:  {:.3f}".format(
        accuracy, precision, recall, f1
    ))
```

```python
classifier = KNeighborsClassifier()
classifier.fit(X_train_std, y_train)
prediction = classifier.predict(X_test_std)
quality(y_test, prediction)
```

```python
# Compute fpr, tpr, thresholds and roc auc
fpr, tpr, thresholds = roc_curve(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)

# Plot ROC curve
plt.figure(figsize=(9,9))
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
```

### Tuning kNN parameters

```python
neighbors = range(1, 50)
f1_score_train = []
f1_score_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_std, y_train)
    f1_score_train.append(f1_score(knn.predict(X_train_std), y_train))
    f1_score_test.append(f1_score(knn.predict(X_test_std), y_test))

plt.figure(figsize=(6,6))
plt.plot(neighbors, f1_score_train, color='blue', label='train')
plt.plot(neighbors, f1_score_test, color='red', label='test')
plt.title("Max test quality: {:.3f}\nBest k: {}".format(max(f1_score_test), np.argmax(f1_score_test)+1))
plt.legend();
```

Таким образом наилучший результат достигается при количестве соседей равном 3.

### Автоматический подбор гиперпараметров

Попробуем еще улучшить нашу модель при помощи автоматического подбора гиперпараметров

```python
%%time
from sklearn.model_selection import GridSearchCV

data = pd.read_csv('data/data.csv')
data = data.drop('Unnamed: 32', axis=1)
diag = data.diagnosis.replace('M', 1).replace('B', 0)
data = data.drop(['id', 'diagnosis'], axis = 1 )

# Разбиваем на train-test выборки в соотношении 70 на 30
X_train, X_test, y_train, y_test = train_test_split(data, diag, test_size=0.3, random_state=37)

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

param_grid = {
        'n_neighbors': range(1, 50),
        'p': range(1, 4)
}

knn = KNeighborsClassifier()

grid_search = GridSearchCV(knn, param_grid, cv=10, scoring='f1')
grid_search.fit(X_train_std, y_train)

print("Best CV score: {:.3f}, best CV k: {}, best CV p: {}".format(
    grid_search.best_score_, grid_search.best_estimator_.n_neighbors, grid_search.best_estimator_.p)
) 

test_predictions = grid_search.best_estimator_.predict(X_test_std)
print("Resulting test score: {:.3f}".format(f1_score(test_predictions, y_test)))
```

```python
from sklearn.model_selection import RandomizedSearchCV

rand_search = RandomizedSearchCV(knn, param_grid, cv=10, scoring='f1', random_state=37)
rand_search.fit(X_train_std, y_train)

print("Best CV score: {:.3f}, best CV k: {}, best CV p: {}".format(
    rand_search.best_score_, rand_search.best_estimator_.n_neighbors, rand_search.best_estimator_.p)
) 

test_predictions = rand_search.best_estimator_.predict(X_test_std)
print("Resulting test score: {:.3f}".format(f1_score(test_predictions, y_test)))
```
