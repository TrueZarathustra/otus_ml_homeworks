## 2. Basic Clustering

### 2.1 Modeling

### 2.1.1 K-means

```python
from sklearn.cluster import KMeans
import matplotlib.ticker as ticker
```

```python
data_r = data.iloc[:,:9].copy(deep=True)
```

```python
data_r.shape
```

```python
from sklearn.metrics import silhouette_score
inertia = []
silhouette = []
k = range(1, 30)
for k_i in k:
    km = KMeans(n_clusters=k_i, n_jobs=-1).fit(data_r)
    inertia.append(km.inertia_)
    if k_i>1:
        silhouette.append(silhouette_score(data_r, km.labels_))
```

```python
fig, ax = plt.subplots()
ax.plot(k, inertia)
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(1, end, 1))
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
k = range(2, 30)
ax.plot(k, silhouette)
ax.xaxis.set_ticks(np.arange(1, 30, 1))
ax.grid()
plt.show()
```

### 2.1.2 Hierarhical

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(data_r, method='centroid', metric='euclidean')
```

```python
fig, ax = plt.subplots(1, 1)
dend = dendrogram(Z, ax=ax, truncate_mode='lastp') 
```

```python
clustering = AgglomerativeClustering().fit(data_r)
```

```python
plt.scatter(data_r.iloc[:,0],data_r.iloc[:,2], c=clustering.labels_, cmap='rainbow')
```

### 2.1.3 DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
def dbscan_demo(eps=0.5, min_pts=1, metric='euclidean'):
    dbscan = DBSCAN(eps=eps, min_samples=min_pts).fit(data_r)
    labels = dbscan.labels_
    plt.figure(figsize=(18, 6))
    plt.scatter(data_r.iloc[:, 0], data_r.iloc[:, 1], c=labels)
    return dbscan

    
# dbs = dbscan_demo(0.75,10)
dbs = dbscan_demo(0.75,10,"cosine")
```

```python
len(dbs.labels_), dbs.labels_.max()
```

### 2.2 Visualizing clusters

### 2.2.1 PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data_r)
principalDf = pd.DataFrame(data = principalComponents, 
                           columns = ['principal component 1',
                                      'principal component 2'])
dbs_labels = pd.Series(dbs.labels_, name='target')
finalDf = pd.concat([principalDf, dbs_labels], axis=1)
```

```python
finalDf.head(5)
```

```python
dbs_labels.value_counts()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1, 2, 3, 4, 5]
colors = ['r', 'g', 'b', 'k', 'y', 'c']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

## 3. Alternative Clustering

### 3.1 Modeling with K-Modes

```python
from kmodes.kmodes import KModes
```

```python
data_adv = data.copy(deep=True)
```

```python
from sklearn.metrics import silhouette_score
silhouette = []
k = range(1, 30)
for k_i in k:
    km = KModes(n_clusters=k_i, n_jobs=-1, init="Cao", n_init=30, max_iter=500).fit(data_adv)
    if k_i>1:
        silhouette.append(silhouette_score(data_adv, km.labels_))
```

```python
fig, ax = plt.subplots()
k = range(2, 30)
ax.plot(k, silhouette)
ax.xaxis.set_ticks(np.arange(1, 30, 1))
ax.grid()
plt.show()
```

```python
km = KModes(n_clusters=3, n_jobs=-1).fit(data_adv)
```

### 3.2 Visualizing clusters

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data_adv)
principalDf = pd.DataFrame(data = principalComponents, 
                           columns = ['principal component 1',
                                      'principal component 2'])
km_labels = pd.Series(km.labels_, name='target')
finalDf = pd.concat([principalDf, km_labels], axis=1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1, 2, 3]
colors = ['r', 'g', 'b', 'k']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
mapper = umap.UMAP().fit(data_adv)
umap.plot.points(mapper, labels=km_labels)
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(data_adv)
```

```python
df_subset = pd.DataFrame()
df_subset['tsne-2d-one'] = tsne_results[:,0]
df_subset['tsne-2d-two'] = tsne_results[:,1]
df_subset['y'] = km.labels_

plt.figure(figsize=(16,10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="y",
    palette=sns.color_palette("bright", 3),
    data=df_subset,
    legend="full",
    alpha=0.8
)
```

## 4. Conclusion

Результат разбиения на кластеры с помощью K-modes показался более интересным.

В результате деления получается 3 кластера.
Их можно интерпретировать следующим образом:

Кластер №0: Люди среднего возраста с собственным домом, преимущественно мужчины. 

Кластер №1: Люди с возрастом немного выше среднего, имеющие большой кредит, преимущественно на автомобиль.

Кластер №2: Люди младшего возраста, не имеющие в собственности недвижость.

Если считать основной задачей кластеризации последующую таргетированную рекламу, то можно сказать, что для разных кластеров подойдет разная реклама, а именно:

Кластер №0: страхование жилья, пенсионные программы, кредиты для бизнеса

Кластер №1: автострахование, пенсионные программы, ипотечные программы

Кластер №2: кредиты на обучение, автокредиты, ипотечные программы

В зависимости от конечно бизнес-задачи, количество кластеров и способ их разбиения может измениться
